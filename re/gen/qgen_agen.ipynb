```python
from loguru import logger

from aimw.app.core.ai_config import get_ai_settings
from aimw.app.resources.icl_templates import icl_gen_templates
from aimw.app.services.batch.batch_qgen_agen_service import QAGenerator
from aimw.app.services.factory.model_factory_service import Factory
from aimw.app.utils import baseline_utils

factory = Factory()

generator = factory.create_llm(model_name="llama3-8b-8192")

qgen_agen = QAGenerator(
    template=icl_gen_templates.template_qgen_agen,
    verbose=True,
    number_of_questions="10",
    model=generator,
    qgen_model_path=get_ai_settings().qgen_model_path,
)
```

```python
r_dir = "./re/data/fiqa/clean/split/"
w_dir = "./re/data/fiqa/clean/base/exp_002/"

generated, failed = baseline_utils.apply_generator(
    qgen_agen, r_dir, w_dir, start_index=150, end_index=151, sleep_time=2
)
```
